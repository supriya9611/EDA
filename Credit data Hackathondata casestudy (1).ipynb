```python
Q1 = credit_data["AMT_INCOME_TOTAL"].describe()['25%']
Q3 = credit_data["AMT_INCOME_TOTAL"].describe()['75%']
IQR = Q3-Q1
UPPER_BOUND = Q3+1.5*IQR
credit_data['AMT_INCOME_TOTAL']=np.where(credit_data['AMT_INCOME_TOTAL']>UPPER_BOUND, UPPER_BOUND, 
                                       credit_data['AMT_INCOME_TOTAL'])
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(credit_data["AMT_INCOME_TOTAL"])
```

# We can see that outliers are removed from the above graph

```python
Q1= credit_data["AMT_CREDIT"].describe()["25%"]
Q3 =credit_data["AMT_CREDIT"].describe()["75%"]
IQR = Q3-Q1
UPPER_BOUND = Q3+1.5*IQR
credit_data["AMT_CREDIT"]= np.where(credit_data["AMT_CREDIT"]>UPPER_BOUND,UPPER_BOUND,credit_data["AMT_CREDIT"])
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(credit_data.AMT_CREDIT)
```

```python
q1=credit_data['AMT_GOODS_PRICE'].describe()['25%']
q3=credit_data['AMT_GOODS_PRICE'].describe()['75%']

iqr=q3-q1
upper_bound=q3+1.5*iqr

credit_data['AMT_GOODS_PRICE']=np.where(credit_data['AMT_GOODS_PRICE']>upper_bound, upper_bound,
                                      credit_data['AMT_GOODS_PRICE'])
```

```python
sns.boxplot(credit_data.AMT_GOODS_PRICE)
plt.show()
```

```python
Q1 = credit_data["CNT_FAM_MEMBERS"].describe()["25%"]
Q3 = credit_data["CNT_FAM_MEMBERS"].describe()["75%"]
IQR = Q3-Q1
UPPER_BOUND = Q3+1.5*IQR
credit_data["CNT_FAM_MEMBERS"]=np.where(credit_data["CNT_FAM_MEMBERS"]>UPPER_BOUND,UPPER_BOUND,credit_data["CNT_FAM_MEMBERS"])
```
